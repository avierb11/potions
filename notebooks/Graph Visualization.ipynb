```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import potions as pt
```

```python
model = pt.HbvModel.default()
graph = model.graph
```

```python
node_labels = {}
for node_id, data in graph.nodes(data=True):
    node_labels[node_id] = data["name"]
```

```python
fig, ax = plt.subplots()

node_options = {
    "node_size": 10_000,
    "node_color": "white",
    "edgecolors": "black",
    "linewidths": 5,
}
edge_options = {
    "width": 5,
    "arrowsize": 20,
}

label_options = {
    "font_size": 16,
}

pos = nx.spring_layout(graph)

nx.draw_networkx_nodes(graph, pos, ax=ax, **node_options)
nx.draw_networkx_edges(graph, pos, ax=ax, **edge_options) # type: ignore
nx.draw_networkx_labels(graph, pos, ax=ax, labels=node_labels, **label_options)  # type: ignore

fig.set_figwidth(8)
fig.set_figheight(6)
```
